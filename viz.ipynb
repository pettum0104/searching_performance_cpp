```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import os

search_times_file = 'results/search_times_ns.csv'
collisions_file = 'results/hash_collisions.csv'
df_times = pd.read_csv(search_times_file)
df_collisions = pd.read_csv(collisions_file)


plt.figure(figsize=(12, 7))

column_labels = {
    'Linear_Search_ns': ' ',
    'BST_Search_ns': 'BST',
    'RBT_Search_ns': 'RBT',
    'HashTable_Search_ns': '-',
    'Multimap_Search_ns': 'std::multimap'
}

for column, label in column_labels.items():
    plt.plot(df_times['Size'], df_times[column], marker='o', linestyle='-', label=label)


plt.xlabel('  ')
plt.ylabel('  ')
plt.title('        ')
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.legend()
plt.tight_layout()

locmaj = ticker.LogLocator(base=10, numticks=12)
plt.gca().xaxis.set_major_locator(locmaj)
locmin = ticker.LogLocator(base=10, subs=(0.1, 0.2, 0.4, 0.6, 0.8), numticks=12)
plt.gca().xaxis.set_minor_locator(locmin)
plt.gca().xaxis.set_minor_formatter(ticker.NullFormatter())

# output_plot_time = 'search_times_comparison.png'
# plt.savefig(output_plot_time)

plt.figure(figsize=(10, 6))

plt.plot(df_collisions['Size'], df_collisions['Collisions'], marker='o', linestyle='-', color='red')
plt.xlabel('  ')
plt.ylabel('    ')
plt.title('    -   ')
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.tight_layout()


plt.gca().xaxis.set_major_locator(locmaj)
plt.gca().xaxis.set_minor_locator(locmin)
plt.gca().xaxis.set_minor_formatter(ticker.NullFormatter())

# output_plot_collisions = 'hash_collisions_plot.png'
# plt.savefig(output_plot_collisions)

plt.show()
```
